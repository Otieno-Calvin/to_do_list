```python
f# the  function below prints the user menu for the to do list.
def show_menu():
    print("HELLO, Welcome to Otienos to-do list \n")
    print("1. add task")
    print("2. task completion")
    print("3. List tasks")
    print("4. delete task")
    print("5. Quit")
    print("\n")

# this function allows the user to input the task to be added to tasks list .
def add_task():
    #user input
    task = input("Enter a task description: \n")
    # adds task to tasks list.
    tasks.append({"description": task, "completed": False})
    # prints success message.
    print("Task added successfully!")

# function marks a task  completed
def mark_completed():
    # user input
    task_number = int(input("Enter the number of the task to mark as completed: "))
    # checks whether the task number is in the range of tasks in list
    if task_number in range(len(tasks)):
    # checks whether the tasks key 'completed' is true else changes completed to true.
        if not tasks[task_number]["completed"]:
            tasks[task_number]["completed"] = True
    # prints success message.
            print("Task marked !! \n completed!")
        else:
            print("task already  completed.")
    # error handling
    else:
        print("Invalid number.")

# function displays tasks list and numbers tasks.
def list_tasks():
    # checks whether tasks list is empty.
    if not tasks:
        print("No tasks in list")
    else:
        print("------TASK LIST--------")
    # iterates through tasks list and prints task description and completion status.
        for index, task in enumerate(tasks):
            completion = "Completed" if task["completed"] else "Not Completed"
            print(f"{index}. {task['description']} - {completion}")
        print("\n")

#function  removes task from list.
def delete_task():
    task_number = int(input("Enter the number of the task to be deleted: "))
    if task_number :
    # checks whether the input is valid , if so , proceed to remove  the specific task using its index in list.
        if  task_number in range(len(tasks)):
            removed_task = tasks.pop(task_number)["description"]
            print(f"'({removed_task})\n deleted")
        else:
            print("Invalid number")
    # error handling.
    else:
        print("Invalid number")

# Function executes specific action depending on users input
def perform_task(action):
    if not action.strip():
        print("No input provided. Please enter a valid option.\n")
    elif action == "1":
        add_task()
    elif action == "2":
        mark_completed()
    elif action == "3":
        list_tasks()
    elif action == "4":
        delete_task()
    elif action == "5":
        print("goodbye!!")
        quit()
    else:
        print("Invalid input. Please try again.\n")

#  a simple to-do list application.

tasks = []

print("Welcome !! \n To-Do List!")

'''
The `while True:` loop  creates an infinite loop that
continuously displays the user menu using the `show_menu()` function and prompts the user to
input an action by using {action} variable.
'''
while True:
    show_menu()
    action = input("what action are you  performing:\n ")
    perform_task(action)
```
